```python
import pandas as pd
```

```python
complaints_data = pd.read_csv("https://data.cityofnewyork.us/resource/erm2-nwe9.csv")
# I was initially thinking of working with the lease agreement rules, from every place, I've lived in but I didn't quite 
# like the text so then I went for 311 complaints. But, I wanted the complaints description, but it wasn't long enough so I decided 
# to work with resolution description.
```

```python
len(complaints_data)
```

```python
# from 1000 resolution descriptions only 20 were unique
len(complaints_data['resolution_description'].unique())
```

```python
lines = complaints_data['resolution_description'].unique()
```

```python
with open("./311_resolutions.txt", "w") as txt_file:
    for line in lines:
        txt_file.write(str(line) + "\n")
```

```python
import sentencex
text = open("./311_resolutions.txt").read()
sentences = sentencex.segment("en", text) # sentencex supports languages other than english! see the docs
# filter empty strings and fix mid-sentence line breaks
sentences = [item.replace("\n", "") for item in sentences if len(item.strip()) > 0] 
sentences = [item for item in sentences if 'nan' not in item.lower()]
```

```python
sentences
```

```python
import random
random.sample(sentences, 10)
```

```python
from shoestrings import TextGenerator
from shoestrings.tokenizers import SimpleWordTokenizer, SimpleCharacterTokenizer
```

```python
gen = TextGenerator(n=2, tokenizer=SimpleWordTokenizer(), source=sentences)
gen.generate_one()
for i in range(8):
    print(gen.generate_one())
    print()
```

```python
# the texts generated started sounding 'syntactically' correct and coherent, meaning that 3-words for the complaint resolutions is enough to 
# create some context
gen = TextGenerator(n=3, tokenizer=SimpleWordTokenizer(), source=sentences)
gen.generate_one()
for i in range(8):
    print(gen.generate_one())
    print()
```
