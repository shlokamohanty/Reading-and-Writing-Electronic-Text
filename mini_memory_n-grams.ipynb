#### consolidation

````python
lines = open("./semantic-categories-2026-04-10.jsonl").readlines()
````

````python
lines[:10]
````

````python
from collections import defaultdict, Counter
import json

# create the dictionary (unknown keys default to an empty list)
cats = defaultdict(list)

for item in lines:  # lines is a list of strings w/encoded JSON objects\
    
    obj = json.loads(item)  # decode the JSON
    
    if len(obj['label']) > 0:
        # optional: merge variations in case (e.g. DAY and day will be counted together)
        lc_text = obj['text'].lower()
        # the extend function appends the contents of one list to another
        cats[lc_text].extend(obj['label'])
````

````python
words_memories = {}
for line in memories:
    words = line.split()
    for word in words:
        if cats[word] != []:
            words_memories[word] = cats[word]
````

````python
# words_memories
````

````python
words_memories_counted = {}
for key, val in words_memories.items():
    words_memories_counted[key] = Counter(val)
````

````python
unique_features = set()
for key, val in cats.items():
    unique_features.update(val)
````

````python
memories_vectorized = {}
for word, counts in words_memories_counted.items():
    vector = [counts[feat] for feat in cols]
    memories_vectorized[word] = vector
````

````python
from simpleneighbors import SimpleNeighbors
````

````python
memories_lookup = SimpleNeighbors(10, 'euclidean')
for name, vec in memories_vectorized.items():
    memories_lookup.add_one(name, vec)
memories_lookup.build()
````

````python
memories_recalled = []
gen = TextGenerator(n=3, tokenizer=SimpleWordTokenizer(), source=memories)
gen.generate_one()
for i in range(10):
    memories_recalled.append(gen.generate_one())

word_recalled_connected = {}
for line in memories_recalled:
    words = line.split()
    for word in words:
        if cats[word] != [] and word in memories_vectorized:
            word_recalled_connected[word] = memories_lookup.nearest(memories_vectorized[word])

words = list(word_recalled_connected.keys())
connected_memory_words = word_recalled_connected[random.choice(words)]
            
connected_memories = []
for line in memories_recalled:
    for word in line.split():
        if word in connected_memory_words:
            connected_memories.append(line)

connected_memories = list(set(connected_memories))

for line in connected_memories:
    print(line + "\n")
````
