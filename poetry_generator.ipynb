```python
import random
import textwrap
```

```python
sal_adj_a = ["dear", "beloved", "sweet", "tender", "kind", "silly"]
sal_adj_b = ["For", "To"]
sal_adjs = ["For", "To", "dear", "beloved", "sweet", "tender", "kind", "silly"]
```

```python
sal_nicknames = ["other half", "dust ocean", "sunshine", "terrapin", 
                 "turtle", "sun", "moon", "wildflower", "stranger",
                "love", "firefly", "dragonfly"]
```

```python
advs = ['lovingly', 'tenderly', 'insufferably', 'longingly',
       'eagerly', 'yearningly', 'wistfully', 'fondly', 'admiringly',
       'affectionately']
```

```python
smells = ["petrichor", "a swelling monsoon", "bergamot", "pinewood", 
          "summer rain", "sea breeze", "honeysuckle", "fresh cut grass",
          "morning dew"]
```

```python
questions_ending_with_nickname = ["How are you doing,", "I wonder how you are,", "Did I ever tell you that many days I hate that I love writing,",
          "Do the words that I weave speak your metric, so are you reading between the lines,", 
                                "Did you ever catch me looking at you,", "Did you ever catch me looking for you,",
                               "How do I see you when you keep passing me by,"]
```

```python
voice = ["soprano", "mezzo-soprano", "contralto", "countertenor", "tenor", "baritone", "bass"]
```

```python
past_tense_sentence1 = ["there was a time when silences felt comforting", 
                       "I remember the moment I developed feelings for you",
                       "when I wrote lyrics just for you", 
                        "I liked calling you by many names"]
```

```python
past_tense_sentence2 = ["and I wish for it again", "I strangely enough feel the same now",
                       "you would have sung them back to me", "and I’ll keep calling you by many more now"]
```

```python
random_phrases1 = ["Well, I’m someone who loves people easily", 
                 "I like the words that your lips move about to say, it’s from the soul and not the head, different than the others",
                 "I think I can read your eyes when they meet mine",
                 "Your soul is warm and kind. In my eyes",
                 "Bright colors look good on you", "Your lips do speak between the lines, and",
                 "Something else pulls me away"]
```

```python
random_phrases2 = ["they become etched in my heart", "you also know cleverly when and how to avoid your eyes from meeting mine",
                  "you’re brighter than how you see yourself", "you could form a verse if you’d like", "you have sucked me in",
                  "but you in your way"]
```

```python
nickname = random.choice(sal_nicknames)
sal_adj = random.choice(sal_adjs)

if sal_adj in sal_adj_b:
    output = sal_adj + " " + "my" + " " + nickname + ",\n"
    output += "\n"
elif sal_adj in sal_adj_a:
    output =  "My" + " " + sal_adj + " " + nickname + ",\n"
    output += "\n"

body = ""

line1 = " ".join(["You smell like", random.choice(smells), "and I miss your scent on me"])
line2 = " ".join([random.choice(questions_ending_with_nickname), "my", nickname])
line3 = " ".join(["I long to hear you sing to me in your", random.choice(["highest", "lowest"]), random.choice(voice)])
line4 = ", ".join(["Many moons ago", random.choice(past_tense_sentence1), random.choice(past_tense_sentence2)])
line5 = ", ".join([random.choice(random_phrases1), random.choice(random_phrases2)])

lines = [line5, line3, line4, line1]
random.shuffle(lines)
for line in lines:
    body += line
    body += ". "

end_nickname = random.choice(sal_nicknames)
for nickname_in_loop in sal_nicknames:
    if nickname == end_nickname:
        if nickname_in_loop != nickname:
            end_nickname = nickname_in_loop
    else:
        continue

body += line2
body += "? "
output += textwrap.fill(body, 60)
output += "\n\nYours " + random.choice(advs) + ",\n"
output += end_nickname
print(output)
```
