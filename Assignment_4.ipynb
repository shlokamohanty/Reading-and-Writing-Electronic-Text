```python
from tracery.modifiers import base_english
```

```python
lines = open("./semantic-categories-2026-04-10.jsonl").readlines()  # read each line into a list of strings
```

```python
from collections import defaultdict, Counter
import json

# create the dictionary (unknown keys default to an empty list)
dictionary = defaultdict(list)

for item in lines:  # lines is a list of strings w/encoded JSON objects\
    
    obj = json.loads(item)  # decode the JSON
    
    if len(obj['label']) > 0:
        # optional: merge variations in case (e.g. DAY and day will be counted together)
        lc_text = obj['text'].lower()
        # the extend function appends the contents of one list to another
        dictionary[lc_text].extend(obj['label'])
```

```python
dictionary['water']
```

```python
dict_counted = {}
for key, val in dictionary.items():
    dict_counted[key] = Counter(val)
```

```python
unique_features = set()
for key, val in dictionary.items():
    unique_features.update(val)
```

```python
# morning
rules = {
    "origin": "#salutation# #desc#\n\n#closings#,\nSun",
    "desc": ["#eyesdesc#"],
    "salutation": ["To", "For"],
    "closings": ["Love", "From", "With love", "Affectionately"],
    
    "recipient_light": ["the light in you", "the gleam in you", "the glow in you"],
    
    "greeting2": ["dear", "love", "dearest", "beloved"],

    "big_synonyms": [key for key, val in dict_counted.items() if val['emotive']], 
    "glistening_synonyms": ["glistening", "gleaming", "glittering", "shining", "shimmering", "glowing", "twinkling", "scintillating"],
    "morning_synonyms": [key for key, val in dict_counted.items() if val['academic']], 
    "sunlight_synonyms": [key for key, val in dict_counted.items() if val['warm']],
    "eye_verbs": ["mirror", "reflect", "embody"],
    "eye_prepositions": ["through", "midst", "amid", "in the thick of", "off", "from"],
    "eye_noun": [key for key, val in dict_counted.items() if val['tangible']],
    "organic": [key for key, val in dict_counted.items() if val['organic']],
    
    
    "eyesdesc": ["#recipient_light#,\n\nYour #organic#, #greeting2#, are so #big_synonyms.s# in the #morning_synonyms.s#, #glistening_synonyms# #eye_prepositions# the #eye_noun#.\nThey #eye_verbs# the #sunlight_synonyms# #glistening_synonyms# #eye_prepositions# your windowpane."],
    
}
grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
grammar.flatten("#origin#")
print(grammar.flatten("#origin#"))
```
